```python
data = {
    'product': ['A', 'A', 'A', 'B', 'B', 'B'],
    'month': ['Jan', 'Feb', 'Mar', 'Jan', 'Feb', 'Mar'],
    'sales': [100, 150, 130, 90, 120, 110]
}
```
